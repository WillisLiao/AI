### Create a random tensor with shape (7, 7)

```python
import torch

tensor1 = torch.rand(7, 7)
tensor1, tensor1.shape
```

### 3. Mat Mul another tensor shape 1, 7 need to transpose

```python
tensor2 = torch.rand(1, 7)
tensor3 = torch.matmul(tensor1, tensor2.T)
tensor3, tensor3.shape
```

### 4. set the random seed to 0 and do 2 and 3 over again

```python
RANDOM_SEED = 0
torch.manual_seed(RANDOM_SEED)
tensorA = torch.rand(7, 7)
torch.manual_seed(RANDOM_SEED)
tensorB = torch.rand(1, 7)

tensorC = torch.matmul(tensorA, tensorB.T)
tensorC, tensorC.shape
```

### 6. Create two random tensors of shape (2, 3) and send them both to the GPU (you'll need access to a GPU for this). Set torch.manual_seed(1234) when creating the tensors (this doesn't have to be the GPU random seed)

```python
# Set random seed
torch.manual_seed(1234)

# Check for access to GPU
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Device: {device}")

# Create two random tensors on GPU
tensor_1 = torch.rand(2, 3).to(device)
tensor_2 = torch.rand(2, 3).to(device)
tensor_1, tensor_2
```

```python
tensor_3 = torch.matmul(tensor_1, tensor_2.T)
tensor_3, tensor_3.shape
```

```python
max = torch.max(tensor_3)
min = torch.min(tensor_3)

max, min
```

```python
arg_max = torch.argmax(tensor_3)
arg_min = torch.argmin(tensor_3)

arg_max, arg_min
```
